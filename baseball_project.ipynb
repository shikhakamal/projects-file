TRAINING PROCESS

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Lasso,LassoCV
```

```python
scaler=StandardScaler()

X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
X1=pd.DataFrame(X_scaled, columns=X.columns)
X1
```

```python
X1.skew()
```

```python
X1.isnull().sum().sum()
```

checking for random state

```python
maxAcc=0
maxRS=0

for i in range(1,500):
    x_train, x_test, y_train, y_test=train_test_split(X1, y, test_size=0.2, random_state=i)
    lr=LinearRegression()
    lr.fit(x_train, y_train)
    y_pred=lr.predict(x_test)
    score=r2_score(y_test, y_pred)
    print('score ', score, 'random state ', i)
    
    if(score>maxAcc):
        maxAcc=score
        maxRS=i
        print('max score is ', maxAcc, 'max random state ', i)
    
```

random state we observe output at random state 450 and score is 91.78%

```python
x_train, x_test, y_train, y_test= train_test_split(X1, y, test_size=0.2, random_state=450)
```

MODEL 1 : LINEAR REGRESSION

```python
lr=LinearRegression()
lr.fit(x_train, y_train)

a=lr.score(x_train, y_train)
b=lr.score(x_test, y_test)
y_pred=lr.predict(x_test)
c=r2_score(y_test, y_pred)

print('training score ', a)
print('testing score ',b)
print('accuracy score ', c)
```

MODEL 2: DECISION TREE REGRESSOR

```python
dtc= DecisionTreeRegressor()
dtc.fit(x_train, y_train)

a=dtc.score(x_train, y_train)
b=dtc.score(x_test, y_test)
y_pred=dtc.predict(x_test)
c=r2_score(y_test, y_pred)

print('training score ', a)
print('testing score ',b)
print('accuracy score ', c)
```

MODEL 3: RANDOM FOREST REGRESSOR

```python
rfr=RandomForestRegressor()
rfr.fit(x_train, y_train)

a=rfr.score(x_train, y_train)
b=rfr.score(x_test, y_test)
y_pred=rfr.predict(x_test)
c=r2_score(y_test, y_pred)

print('training score ', a)
print('testing score ',b)
print('accuracy score ', c)
```

MODEL 4: SVM

```python
svc=SVR()
svc.fit(x_train, y_train)

a=svc.score(x_train, y_train)
b=svc.score(x_test, y_test)
y_pred=svc.predict(x_test)
c=r2_score(y_test, y_pred)

print('training score ', a)
print('testing score ',b)
print('accuracy score ', c)
```

MODEL 5: LASSO CV

```python


lasscv=LassoCV(alphas=None, cv=7, max_iter=1000)
lasscv.fit(x_train, y_train)

alpha=lasscv.alpha_
alpha
```

```python
lass_reg=Lasso(alpha)
lass_reg.fit(x_train, y_train)

a=lass_reg.score(x_train, y_train)
b=lass_reg.score(x_test, y_test)
y_pred=lass_reg.predict(x_test)
c=r2_score(y_test, y_pred)

print('training score ', a)
print('testing score ',b)
print('accuracy score ', c)
```
