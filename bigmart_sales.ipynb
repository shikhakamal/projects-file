LABEL ENCODER

many columns are of object type and it contains string values, so machine is not able to understand the value of string so we need to convert string values into numerical one by using label encoder.

```python
df.columns
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc=OrdinalEncoder()
```

```python
for i in df.columns:
    if df[i].dtypes=='object':
        df[i]=enc.fit_transform(df[i].values.reshape(-1,1))
        
```

```python
df
```

```python
df.dtypes
```

FIND NAN VALUES AND FILL THESE NAN VALUES WITH MEAN IN CONTINUOS DATA AND WITH MODE IN CATAGORICAL DATA

now all string is being converted into numerical one, now we check for null spaces and then fill the nan values by using iterative imputer as it is most effective one.

```python
df.isnull().sum()
```

```python
np.mean(df['Item_Weight'])

```

```python
df['Item_Weight']=df['Item_Weight'].fillna(np.mean(df['Item_Weight']))
df['Item_Weight']
```

```python
import statistics

df['Outlet_Size']=df['Outlet_Size'].fillna(statistics.mode(df['Outlet_Size']))
df['Outlet_Size']
```

```python
df['Item_Outlet_Sales']=df['Item_Outlet_Sales'].fillna(np.mean(df['Item_Outlet_Sales']))
df['Item_Outlet_Sales']
```

```python
df.isnull().sum()
```

null values are being removerd and filled by its mean and mode values in both trained and testing data sets


```python
df.describe()
```

the counts of both trained and testing datasets showing equal counts , which we may say that it is not having any null spaces presnt now. So we can proceed with this.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, accuracy_score

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
X=df_new.drop('Item_Outlet_Sales', axis=1)
y=df_new['Item_Outlet_Sales']
```

```python
X
```

```python
X.dtypes
```

```python
y
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.20, random_state=25)
```

MODEL 1: LINEAR REGRESSION MODEL

```python
lr=LinearRegression()
lr.fit(X_train, y_train)

a=lr.score(X_train, y_train)
y_pred=lr.predict(X_test)
#b=r2_score(y_train, y_pred)

print('the training score is:',a)
#print('the r2 score is:',b)
```

```python
dtc=DecisionTreeRegressor()
dtc.fit(X_train, y_train)

a =dtc.score(X_train, y_train)
print('the training score is :', a)

y_pred=dtc.predict(X_test)
#b=r2_score(y, y_pred)
#b
```

```python
rfr=RandomForestRegressor()
rfr.fit(X_train, y_train)

a=rfr.score(X_train, y_train)
a
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
y_pred=knn.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
```
