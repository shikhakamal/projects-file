TRAINING PROCESS:

Standard scaler used

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, accuracy_score

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV , RandomizedSearchCV
```

```python
x
```

```python
y
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(x)
```

```python

X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=25)
```

MODEL 1: LINEAR REGRESSION

```python
lr=LinearRegression()
lr.fit(X_train, y_train)
a=lr.score(X_train, y_train)
b=lr.score(X_test, y_test)

y_pred=lr.predict(X_test)
c=r2_score(y_test, y_pred)


print('Results for linear regression: \n')
print('The score of trained data is ',a)
print('The score of test data is ', b)
print('Performace is ',c)
```

MODEL 2: DECISION TREE REGRESSOR

```python
dtr=DecisionTreeRegressor()
dtr.fit(X_train, y_train)

a=dtr.score(X_train, y_train)
b=dtr.score(X_test, y_test)

y_pred=dtr.predict(X_test)
c=r2_score(y_test, y_pred)


print('Results for decision tree regressor: \n')
print('The score of trained data is ',a)
print('The score of test data is ', b)
print('Performace is ',c)
```

MODEL 3: RANDOM FOREST REGRESSOR

```python
rfr=RandomForestRegressor()
rfr.fit(X_train, y_train)

a=rfr.score(X_train, y_train)
b=rfr.score(X_test, y_test)

y_pred=rfr.predict(X_test)
c=r2_score(y_test, y_pred)

print('Results for decision tree regressor: \n')
print('The score of trained data is ',a)
print('The score of test data is ', b)
print('Performace is ',c)
```

MODEL 4: LASSO REGRESSION

```python
from sklearn.linear_model import Lasso,LassoCV, Ridge,RidgeCV
```

```python
lasscv=LassoCV(alphas=None, cv=10, max_iter=1000)
lasscv.fit(X_train, y_train)
```

```python
alpha=lasscv.alpha_
alpha
```

```python
lasso_reg=Lasso(alpha)
lasso_reg.fit(X_train, y_train)
```

```python
b=lasso_reg.score(X_test, y_test)
a=lasso_reg.score(X_train, y_train)

y_pred=lasso_reg.predict(X_test)
c=r2_score(y_test, y_pred)

print('Results for lasso regression: \n')
print('The score of trained data is ',a)
print('The score of test data is ', b)
print('Performace is ',c)
```

MODEL 5: RIDGE REGRESSION


```python
alphas=np.random.uniform(low=0, high=10, size=(50))
ridgecv=RidgeCV(alphas=alphas, cv=10)
ridgecv.fit(X_train, y_train)
ridgecv.alpha_
```

```python
ridge_reg=Ridge(alpha=ridgecv.alpha_)
ridge_reg.fit(X_train, y_train)

a=ridge_reg.score(X_train, y_train)
b=ridge_reg.score(X_test, y_test)

y_pred=ridge_reg.predict(X_test)
c=r2_score(y_test, y_pred)

print('Results for Ridge regression: \n')
print('The score of trained data is ',a)
print('The score of test data is ', b)
print('Performace is ',c)
```

MODEL 6: SVM(SUPPORT VECTOR MACHINE)

```python
from sklearn.svm import SVR
svr=SVR()

svr.fit(X_train, y_train)
a=svr.score(X_train, y_train)
b=svr.score(X_test, y_test)

y_pred=svr.predict(X_test)
c=r2_score(y_test, y_pred)

print('Results for SVR: \n')
print('The score of trained data is ',a)
print('The score of test data is ', b)
print('Performace is ',c)
```
