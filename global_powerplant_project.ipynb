```python
imp=SimpleImputer()

df['latitude']=imp.fit_transform(df['latitude'].values.reshape(-1,1))
df['longitude']=imp.fit_transform(df['longitude'].values.reshape(-1,1))
df['generation_gwh_2014']=imp.fit_transform(df['generation_gwh_2014'].values.reshape(-1,1))
df['generation_gwh_2015']=imp.fit_transform(df['generation_gwh_2015'].values.reshape(-1,1))
df['generation_gwh_2016']=imp.fit_transform(df['generation_gwh_2016'].values.reshape(-1,1))
df['generation_gwh_2017']=imp.fit_transform(df['generation_gwh_2017'].values.reshape(-1,1))
df['generation_gwh_2018']=imp.fit_transform(df['generation_gwh_2018'].values.reshape(-1,1))
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
list=['other_fuel1', 'owner', 'generation_data_source']
for i in list:
    df[i]=le.fit_transform(df[i])
```

```python
df.dtypes
```

```python
df['other_fuel1']=df['other_fuel1'].fillna(stats.mode(df['other_fuel1']))
df['owner']=df['owner'].fillna(stats.mode(df['owner']))
df['geolocation_source']=df['geolocation_source'].fillna(stats.mode(df['geolocation_source']))
df['generation_data_source']=df['generation_data_source'].fillna(stats.mode(df['generation_data_source']))
df['commissioning_year']=df['commissioning_year'].fillna(stats.mode(df['commissioning_year']))
df['year_of_capacity_data']=df['year_of_capacity_data'].fillna(stats.mode(df['year_of_capacity_data']))
```

TRAINING PROCESS: PRIMARY FUEL AS TARGET VARIABLE

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, r2_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
```

to check the random state

```python
maxAcc=0
maxRS=0

dt=DecisionTreeClassifier()
for i in range(1,500):
    x_train, x_test, y_train, y_test=train_test_split(X_new, y, test_size=0.2, random_state=i)
    dt.fit(x_train, y_train)
    y_pred=dt.predict(x_test)
    score=accuracy_score(y_test, y_pred)
    print('score ', score, 'random state ', i)
    
    if(score>maxAcc):
        maxAcc=score
        maxRS=i
        print('max score is ',maxAcc, 'at random state ', i)
    
    
```

max score is 0.93% at random state 424

```python
x_train, x_test, y_train, y_test=train_test_split(X_new, y, test_size=0.2, random_state=424)
```

MODEL 1: DECISION TREE CLASSIFIER

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train, y_train)

a=dtc.score(x_train, y_train)
b=dtc.score(x_test, y_test)
y_pred=dtc.predict(x_test)
c=accuracy_score(y_test, y_pred)

print('training score is ', a)
print('testing score is ', b)
print('accuracy score is ', c)
print('confusion matrix \n', confusion_matrix(y_test, y_pred))
print('classification report \n', classification_report(y_test, y_pred))
```

MODEL 2: RANDOM FOREST CLASSIFIER

```python
rfc=RandomForestClassifier()
rfc.fit(x_train, y_train)

a=rfc.score(x_train, y_train)
b=rfc.score(x_test, y_test)
y_pred=rfc.predict(x_test)
c=accuracy_score(y_test, y_pred)

print('training score is ', a)
print('testing score is ', b)
print('accuracy score is ', c)
print('confusion matrix \n', confusion_matrix(y_test, y_pred))
print('classification report \n', classification_report(y_test, y_pred))
```

MODEL 3: SVM (SUPPORT VECTOR MACHINES )

```python
svc=SVC()
svc.fit(x_train, y_train)

a=svc.score(x_train, y_train)
b=svc.score(x_test, y_test)
y_pred=svc.predict(x_test)
c=accuracy_score(y_test, y_pred)

print('training score is ', a)
print('testing score is ', b)
print('accuracy score is ', c)
print('confusion matrix \n', confusion_matrix(y_test, y_pred))
print('classification report \n', classification_report(y_test, y_pred))
```

MODEL 4: K NEIGHBORS CLASSIFIER(KNN)

```python
knn=KNeighborsClassifier()
knn.fit(x_train, y_train)

a=knn.score(x_train, y_train)
b=knn.score(x_test, y_test)
y_pred=knn.predict(x_test)
c=accuracy_score(y_test, y_pred)

print('training score is ', a)
print('testing score is ', b)
print('accuracy score is ', c)
print('confusion matrix \n', confusion_matrix(y_test, y_pred))
print('classification report \n', classification_report(y_test, y_pred))
```
