TRAINING PROCESS

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
df
```

```python
x=df.drop('Churn', axis=1)
x
```

```python
y=df['Churn']
y
```

checking for random state

```python
maxAcc=0
maxRS=0

for i in range(1,300):
    lr=LogisticRegression()
    x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.20, random_state=i)
    lr.fit(x_train, y_train)
    y_pred=lr.predict(x_test)
    score=accuracy_score(y_test, y_pred)
    print('score is ',score, 'at random state ',i)
    
    if score>maxAcc:
        maxAcc=score
        maxRS=i
        print('max score is ', maxAcc, 'at random state ',maxRS)
```

max score is 82.469 % at random state 247

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.20, random_state=247)
```

RUN ALL MODELS

```python
lr=LogisticRegression()
gnb=GaussianNB()
dtc=DecisionTreeClassifier()
rfc=RandomForestClassifier()
knn=KNeighborsClassifier()

list=[lr, gnb, dtc, rfc, knn]
for i in list:
    print('\n\n The ',i,'is :\n')
    i.fit(x_train, y_train)
    y_pred=i.predict(x_test)
    print('accuracy score ', accuracy_score(y_test, y_pred))
    print('confusion matrix \n', confusion_matrix(y_test, y_pred))
    print('classification_matrix \n', classification_report(y_test, y_pred))
```

the best results is shown in logistic regression

CROSS VALIDATION SCORE

```python
from sklearn.model_selection import cross_val_score
```

```python
for i in list:
    scr=cross_val_score(i, x, y, cv=7)
    print('\n\n The cross validation score of ',i,'is: ', scr.mean())
```

random forest and logistic regression is giving best results, hence we continue with logistic regression as it is giving best results in cross validation as well as training model

ROC AUC CURVE

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
y_pred1=lr.predict(x_test)

fpr, tpr, thresholds= roc_curve(y_test, y_pred1)
print('\n fpr', fpr)
print('\n tpr', tpr)
print('\n thresholds', thresholds)
```

```python
plt.plot([0,1], [0,1], 'k--')
plt.title('Logistic Regression roc auc curve')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.show()

auc_score=roc_auc_score(y_test, y_pred1)
print('roc auc score is ', auc_score)
```
