```python
rank_country_drop=data_modify.drop(['Country','Happiness Rank'], axis=1)
rank_country_drop
```

```python
happy=rank_country_drop
```

applying linear regression model to predict happiness score as happiness score is dependent variable and continous data.


```python
from sklearn.linear_model import LinearRegression
X=happy.drop('Happiness Score', axis=1)
y=happy['Happiness Score']
lr=LinearRegression()
lr.fit(X,y)
```

```python
y_pred=lr.predict(X)
y_pred
```

```python
sns.scatterplot(y=happy['Happiness Score'], x=y_pred)
plt.xlabel('Pedicted happiness')
plt.ylabel('happiness score')
plt.title('Graph between happiness and predicted happiness')
```

```python
msehappy=np.mean((happy['Happiness Score']-y_pred)**2)
msehappy
```

```python
lr2=LinearRegression()
lr2.fit(X[['Dystopia Residual']], happy['Happiness Score'])
```

```python
y_pred2=lr2.predict(X[['Dystopia Residual']])
y_pred2
```

```python
msefamily=np.mean((happy['Happiness Score']-y_pred2 )**2)
msefamily
```
